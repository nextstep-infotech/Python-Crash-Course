# Break and Continue Statement in Python

## Break Statement

The Python break statement immediately terminates a loop entirely. Program execution proceeds to the first statement following the loop body.

### Break in While Loop

while (condition1):<br>
  &emsp;statement(s)<br>
  &emsp;if (condition2):<br>
    &emsp;&emsp;break

```python
#Example:
i = 1
while i<=10:
    print(i)
    if i==5:
        break
    i += 1
```

### Break in For Loop

for var in sequence:<br>
  &emsp;statement(s)<br>
  &emsp;if (test_expression2):<br>
    &emsp;&emsp;break

```python
#Example:
for i in range(2, 10):
    print(i)
    if i==5:
        break
```

## Print results for equation x**2 + 5x + 10 in while loop by incrementing value of x by 2 and if  results exceeds 100, break the loop.

```python
#Solution
x = 0
res = 0
while True:
    res = x**2 + 5*x + 10
    if res > 100:
        break
    print(res)
    x += 2
```

## Check if there is 'i' in 'string' using for loop and if found, break the loop.

```python
#Solution
for s in 'string':
    if s == "i":
        break
    print(s)
```

## Continue Statement

The Python continue statement immediately terminates the current loop iteration. Execution jumps to the top of the loop, and the controlling expression is re-evaluated to determine whether the loop will execute again or terminate.

### Continue in While Loop

while (condition1):<br>
  &emsp;statement(s)<br>
  &emsp;if (condition2):<br>
    &emsp;&emsp;continue

```python
#Example:
i = 0
while i<=10:
    if i == 5:
        i += 1
        continue
    else:
        print(i)
    i += 1
```

### Continue in For Loop

for var in sequence:<br>
  &emsp;statement(s)<br>
  &emsp;if (condition2):<br>
    &emsp;&emsp;continue

```python
#Example:

for i in range(1, 15, 3):
    if i==7:
        continue
    else:
        print(i)
```

## Skip value 5 when printing from 10 to 1 using while loop with continue statement.

```python
#Solution:
i = 10
while i >= 1:
    if i == 5:
        i -= 1
        continue
    else:
        print(i)
    i -= 1
```

## Write a Python program that prints all the numbers from 0 to 10 except numbers divisible by 3.

```python
#Solution
for i in range(11):
    if i%3 == 0:
        continue
    else:
        print(i)
```

## Write a Python program to guess a number between 1 to 9.

```python
while True:
    num = int(input("Enter your guess number: "))
    if num>1 and num<9:
        print("Great. Your guess number lies between 1 and 9.")
        break
```

## Take integer inputs from user until he/she presses q ( Ask to press q to quit after every integer input ). Print average and product of all numbers.

```python
total = 0
counter = 0
product = 1
while True:
    user_input =  input("Enter your number or press q to quit: ")
    if user_input == "q":
        break
    total = total + int(user_input)
    counter += 1
    product = product * int(user_input)
print("Average: ", total/counter)
print("Product: ", product)
    
```
