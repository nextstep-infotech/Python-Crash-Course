# Sys Module in Python

The sys module is a built in module in python.

It provides several functions and variables to manipulate the python runtime environment.

To use its functions we have to import it.

The commonly used functions and variables of sys module are as follows:

```python
import sys
```

## Version

It is used to get string containing the version of Python Interpreter with some additional information.

```python
print(sys.version)
```

## getrecursionlimit()

sys.getrecursionlimit() method is used to find the current recursion limit of the interpreter or to find the maximum depth of the Python interpreter stack.

```python
print(sys.getrecursionlimit())
```

```python
def fact(n):
    if n==0:
        return 1
    else:
        return n*fact(n-1)

f = int(input("Enter the number: \n"))
print(fact(f))
```

## setrecursioinlimit()

sys.setrecursionlimit() method is used to set the maximum depth of the Python interpreter stack to the required limit.

```python
sys.setrecursionlimit(1000)
```

## exit()

sys.exit([arg]) can be used to exit the program. 

The optional argument arg can be an integer giving the exit or another type of object. 

If it is an integer, zero is considered “successful termination”.

Note: A string can also be passed to the sys.exit() method.

```python
print("Hello world")
sys.exit(0)
print("Bye world")
```

```python
age_limit = 15

if age_limit<18:
    sys.exit("Age less than 18 is not allowed to vote.")
else:
    print("Have you voted?")
```

## platform

This variable is used to identify the host operating system which we are using.

```python
print(sys.platform)
```

## executable

This specifies the path of python executable file.

```python
print(sys.executable)
```

## modules

Returns the list of available modules.

```python
print(sys.modules)
```

```python
for m in sys.modules:
    print(m)
    print()
```

## argv

This is actually a list which stores the command line arguments.

Command-line arguments are those which are passed during the calling of the program along with the calling statement. 

To achieve this using the sys module, the sys module provides a variable called sys.argv. It’s main purpose are:

+ It is a list of command-line arguments.
+ len(sys.argv) provides the number of command-line arguments.
+ sys.argv[0] is the name of the current Python script.

The command line arguments or CLA are the parameters which are passed from command prompt while executing a python program.

They are similar to command line arguments of C/C++/Java.
